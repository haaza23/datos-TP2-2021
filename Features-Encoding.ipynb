```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelBinarizer
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
```

```python
train_values = pd.read_csv('data/train_values.csv')
train_labels = pd.read_csv('data/train_labels.csv')
test_values = pd.read_csv('data/test_values.csv')
```

```python
test_values.head()
```

<h1>Label encoding</h2>

```python
def LabelEncoding_categorical(df):
    categorical_features = ['land_surface_condition', 'foundation_type', 'roof_type',
                            'ground_floor_type', 'other_floor_type', 'position',
                            'plan_configuration','legal_ownership_status']
    for features in categorical_features:
        le = LabelEncoder()
        df[features] = le.fit_transform(df[features].astype(str))
    return df

df = LabelEncoding_categorical(train_values)
df_test = LabelEncoding_categorical(test_values)
```

```python
train_values = train_values[['building_id']]
train_values = pd.merge(train_values,df, on='building_id', how='inner') 
train_values.to_csv("data/train_values_enc.csv", index=False)
```

```python
test_values = test_values[['building_id']]
test_values = pd.merge(test_values, df_test, on='building_id', how='inner') 
test_values.to_csv("data/test_label_enc.csv", index=False)
```
